**Importing Libraries**

```python
import pandas as pd               # For data handling
import seaborn as sns             # For visualization and loading the Iris dataset
import matplotlib.pyplot as plt   # For plotting graphs
```

 **Loading the iris Dataset**

```python
df = sns.load_dataset('iris')
```

**Visualizing Scatter Plot â€” Relationship Between Two Variables**

```python
# Scatter plot to see relationship between Sepal Length and Sepal Width
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
# Set title for the plot
plt.title("Sepal Length vs Sepal Width by Species")
# Show the plot
plt.show()
```

The scatter plot shows the relationship between **sepal length** and **sepal width** for each iris species. We can observe that:
- **Setosa** has generally higher sepal width and lower sepal length.
- **Versicolor** and **Virginica** overlap slightly but can still be visually separated.
- This plot helps in understanding how well these two features separate the species.


**Visualizing Histogram Plot  â€” Distribution of a Single Variable**

```python
# Histogram to show distribution of Petal Length
sns.histplot(data=df, x='petal_length', kde=True)  # kde=True adds a smooth curve
# Set title for the plot
plt.title("Distribution of Petal Length")
# Show the plot
plt.show()
```

The histogram displays the distribution of **petal length** across all iris flowers.
- There are three clear peaks, which may correspond to the three species.
- **Petal length** is a strong distinguishing feature; Setosa has much smaller petal lengths.
- The KDE curve shows the smooth probability distribution.


**KDE(Kernel Density Estimate):**
        Itâ€™s a smooth line that shows the shape of the data distribution.
Think of it as a smoothed-out version of a histogram instead of bars, it uses a curve to show where values are more or less common.

**Visualizing Box Plot â€” Detecting Outliers and Spread**

```python
# Box plot to show spread and outliers in Petal Width by species
sns.boxplot(data=df, x='species', y='petal_width')
# Set title for the plot
plt.title("Box Plot of Petal Width by Species")
# Show the plot
plt.show()
```

The box plot compares the spread and outliers of **petal width** across the three species:
- **Setosa** has the smallest petal width with no overlap with the other species.
- **Virginica** shows the largest petal width and more variation.
- **Versicolor** falls between Setosa and Virginica.
